```python
#Step1 : Load CSV
#FILE_PATH = 'EmergencyTriageDataset_Reduced_Dirty.csv' for manual file upload
FILE_PATH = '/content/drive/MyDrive/Colab Notebooks/Carisurg/Copy of EmergencyTriageDataset_Reduced_Dirty.csv'

df_raw = pd.read_csv(FILE_PATH)
```

```python
# Step 2: See unique values
print(f"Dataset loaded: {df_raw.shape[0]} rows x {df_raw.shape[1]} columns")
print(f"Columns: {list(df_raw.columns)}")
```

```python
# Step 3: Count how many of each value we have
print("Gender value counts:")
print(df_raw['Gender'].value_counts())
```

```python
# Step 4: Remap values
gender_map = {
    'Male':   1,
    'MALE':   1,
    '1':      1,
    'Female': 0,
    'FEMALE': 0,
    '0':      0
}

df_raw['Gender_Clean'] = df_raw['Gender'].map(gender_map)

print("After cleaning:")
print(df_raw['Gender_Clean'].value_counts())
print(f"\nAny NaN values? {df_raw['Gender_Clean'].isnull().sum()}")
```

```python
# Step 5: Drop column
df_raw = df_raw.drop(columns=['Gender'])
```

```python
# Step 6: Rename column
df_raw = df_raw.rename(columns={'Gender_Clean': 'Gender'})

print("Column 'Gender' after cleaning:")
print(df_raw['Gender'].value_counts())
print(f"Data type: {df_raw['Gender'].dtype}")
df_raw.head()
```
